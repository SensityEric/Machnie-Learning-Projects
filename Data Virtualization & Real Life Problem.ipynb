### Load FIFA Data

```python
fifa = pd.read_csv("fifa_data.csv")
fifa.head()
```

## Histograms

```python
plt.figure(figsize=(8,6), dpi=300)
bins =[40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins=bins, color="red")



plt.title("Distribution of Players in Fifa",fontdict={'fontname' :
'Poppins','fontsize':15,'fontweight': 'bold'})
plt.xticks(bins)
plt.ylabel('Number of players')
plt.xlabel('Skill Level')
plt.show()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == "Left"].count()[0]
left

right = fifa.loc[fifa['Preferred Foot'] == "Right"].count()[0]
labels = ['left',"right"]
colors = ["Yellow",'blue']
plt.figure(figsize =(8,6), dpi= 300)
plt.title("Preferred Foot of Footballers",fontdict={'fontname' :
'Poppins','fontsize':15,'fontweight': 'bold'})

plt.pie([right,left], labels =labels,colors = colors, autopct='%.2f %%')
plt.savefig('pie.png',dpi = 300)
plt.show()
```
