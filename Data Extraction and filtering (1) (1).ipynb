# String Accessor

```python
G_country = gdata[gdata.country.str.startswith('G')]  # derives countries that startswith a "G"
G_country
```

# Contains and Endswith

```python
G_country = gdata[gdata.country.str.contains('Gh')]  # derives countries that contains "Gh"
G_country
```

```python
G_country = gdata[gdata.country.str.endswith('s')]  # derives countries that endswith "s"
G_country
```

# Tilde operator ~

```python
G_country = gdata[~gdata.country.str.endswith('s')]  # derives countries that endswith "s"
G_country                            # gives the opposite of the logic
```
