```python
#Print histogram for each column. visualization of distribution, value range, scale..
life_data.hist(bins=22, figsize=(12,14))
plt.savefig('life_data22.png',dpi=300)
plt.show()
plt.subplots_adjust(hspace=1)
```

```python
plt.figure(figsize=(10,12), dpi=300)
plt.title("Life expectancy Vs. Schooling")
plt.scatter(life_data["Schooling"], life_data['Life expectancy '])
plt.ylabel('Life Expectancy')
plt.xlabel('Schooling')
plt.xticks([10,20,30])
plt.show()

plt.figure(figsize=(10,12), dpi=300)
plt.title("Life expectancy Vs. Alcohol")
plt.scatter(life_data["Alcohol"], life_data['Life expectancy '])
plt.ylabel('Life Expectancy')
plt.xlabel('Alcohol')
plt.xticks([10,20,30])
plt.show()
```
