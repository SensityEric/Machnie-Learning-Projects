```python
# import libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
#load Data
df_golf = pd.read_csv("C:/Users/Admin/Downloads/golfdataset.csv")
```

```python
df_golf
```

```python
 #encode data : string to numeric
from sklearn import preprocessing 
le = preprocessing.LabelEncoder()
df_golf["Outlook"] = le.fit_transform(df_golf.Outlook.values)
df_golf["Temp"] = le.fit_transform(df_golf.Temp.values)
df_golf["Humidity"] = le.fit_transform(df_golf.Humidity.values)
df_golf["Windy"] = le.fit_transform(df_golf.Windy.values)
df_golf["PlayGolf"] = le.fit_transform(df_golf.PlayGolf.values)
df_golf
```

```python
# splitting data into train and test
labels = df_golf["PlayGolf"]
features = df_golf.drop('PlayGolf', axis = 1)

X_train , X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2,random_state=0)
```

```python
# training predictions on the the testing set
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
y_test
```

```python
#making predictions on the testing set
y_pred = gnb.predict(X_test)
```

```python
y_pred
```

```python
#comparing the actual response (y_test) with predicted response
from sklearn import metrics 
print("Gaussian Naive Bayes Model accuracy (in%) :)" , metrics.accuracy_score(y_test,y_pred)*100)
```
